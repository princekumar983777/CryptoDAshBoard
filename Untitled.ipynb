```python
import json

def convert_blocks_to_table(blocks):
    all_rows = []

    for block in blocks:
        ts = block[0]["timestamp"]
        raw = block[0]["raw_data"]

        try:
            # Convert raw text "[{...}, {...}]" to python list
            data_list = json.loads(raw)

            # Add timestamp to each dictionary
            for item in data_list:
                item["timestamp"] = ts
                all_rows.append(item)

        except Exception as e:
            print("Error parsing block:", e)
            print("Raw block:", raw[:120])

    # Convert final list â†’ DataFrame
    df = pd.DataFrame(all_rows)
    return df
```
